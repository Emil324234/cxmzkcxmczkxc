Цель: Использовать датасет с прошлой практической работы и визуализировать данные в виде графиков

```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np

DS3_weapon = pd.read_csv(r'E:\OneDrive\Рабочий стол\DS3_weapon.csv')
plt.plot(DS3_weapon.columns, DS3_weapon.nunique(), label = "Уникальное значение")
plt.title("Соотношение столбцов")
plt.xlabel("Столбец")
plt.ylabel("Особое значение")
plt.legend()
plt.show()
```

Здесь я получил линейчатый график на основе названия столбца и кол-ва уникальных значений в нём через библиотеку matplotlib и pandas. 

```python
plt.bar(DS3_weapon.columns, DS3_weapon.nunique(), label = "Уникальное значение")
plt.title("Соотношение столбцов")
plt.xlabel("Столбец")
plt.ylabel ("Уникальное значение")
plt.legend()
plt.show()
```

Здесь я получил столбчатую диаграмму на основе названия столбца и кол-ва уникальных значений в них через те же библиотеки и с теми же результатами, что и в линейчатом графике до этого.

```python
plt.plot(DS3_weapon.columns, DS3_weapon.nunique(), label = "Уникальное значение", color = "yellow")
plt.bar(DS3_weapon.columns, DS3_weapon.nunique(), label = "Уникальное значение", color = "black")
plt.title("Соотношение столбцов")
plt.xlabel("Столбец")
plt.xlabel("Уникальное значение")
plt.legend()
plt.show()
```

Здесь я наложил линейный график на столбчатую диаграмму на основе названия столбцов и количества уникальных значений в них.

```python
plt.plot(DS3_weapon["Name"].tail(130), DS3_weapon.nunique(axis = 1).tail(130), label = "Уникальное значение в последних 130 оружиях", color = "yellow")
plt.bar(DS3_weapon["Name"].tail(130), DS3_weapon.nunique(axis = 1).tail(130), label = "Уникальное значение в последних 130 оружиях столбцы", color = "black")
plt.title("Уникальное значения в последних 130 оружиях")
plt.xlabel("Имя оружие")
plt.ylabel("Кол-во уникальных значений")
plt.legend()
plt.show()
```

Здесь последняя столбчатая диаграмма и линейный график, созданные на основе имени последних 130 оружиях и уникальных значений у их строк.

```python
weaponcount = DS3_weapon["Category"].value_counts()
plt.pie(weaponcount, labels=weaponcount.index, autopct="%1.1f%%", colors=["yellow","black"])
plt.title("Соотношение по оружию")
plt.show()
```

Эта часть кода отвечает за круговую диаграмму, созданную на основе соотношения оружий по категориям.

```python
nameweaponcount = DS3_weapon["Name"].value_counts()
plt.pie(nameweaponcount, labels=nameweaponcount.index, autopct="%1.1f%%", colors=["black","yellow","blue","orange","red","purple","orange","brown"])
plt.title("Соотношение оружия по названию")
plt.show()
```

Эта часть кода отвечает за круговую диаграмму, созданную на основе соотношения оружия по именам.

```python
plt.scatter(DS3_weapon["Name"].head(10), DS3_weapon["Category"].head(10), color = "yellow")
plt.title("Связь между именем оружия и категорией")
plt.xlabel("Название оружия")
plt.ylabel("Категории")
plt.show()
```

Эта часть кода отвечает за создание диаграммы рассеяния на основе связи между названием оружия и категории первых 10 оружий.

```python
plt.scatter(DS3_weapon["Category"].head(20), DS3_weapon["Name"].head(20), color = "yellow")
plt.title("Связь между категорией оружия и первым названием оружия")
plt.xlabel("Категория")
plt.ylabel("Названия оружия")
plt.show()
```

Эта часть кода отвечает за создание диаграммы рассеяния на основе связи между одним названием оружия и категорией первых 20 оружий.

```python
plt.hist(DS3_weapon["Damage"].head(30), bins=80, color="yellow", edgecolor="black")
plt.title("Распределение оружий по урону")
plt.xlabel("Урон")
plt.ylabel("Оружие")
plt.show()
```

Этот кусок кода отвечает за создание гистраграммы на основе распределения урона оружия первых 30 оружий.

```python
plt.hist(DS3_weapon["Name"], bins=80, color="yellow", edgecolor="black")
plt.title("Распределение всего урона по оружиям")
plt.xlabel("Урон")
plt.ylabel("Оружие")
plt.show()
```

Этот кусок кода отвечает за создание гистраграммы на основе распределения урона. 

```python
plt.hist(DS3_weapon["Stat Requirements"], bins=80, color="yellow", edgecolor="black")
plt.title("Распределение статов")
plt.xlabel("Какие статы")
plt.ylabel("Оружие")
plt.show()
```
